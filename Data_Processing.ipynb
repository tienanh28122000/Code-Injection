## Prepare and format data

```python
# Concat http_method with http_request and http_content
```

```python
def create_text(method, request, body):
    if method == 'GET':
        return str(method) + ' ' + str(request)
    return str(method) + ' ' + str(request) + ' ' + str(body)
            
df_capec['text'] = df_capec.apply(lambda x: create_text(x['request_http_method'], x['request_http_request'], x['request_body']), axis = 1)
```

```python
df_capec.head()
```

```python
# Mapping values for classification
```

```python
def count_duplicate(a0,a1,a2,a3,a4,a5,a6,a7,a8,a9,a10,a11,a12,a13):
    if a0 == 1:
        return '000 - Normal'
    elif a1 == 1:
        return '272 - Protocol Manipulation'
    elif a2 == 1:
        return '242 - Code Injection'
    elif a3 == 1:
        return '88 - OS Command Injection'
    elif a4 == 1:
        return '126 - Path Traversal'
    elif a5 == 1:
        return '66 - SQL Injection'
    elif a6 == 1:
        return '16 - Dictionary-based Password Attack'
    elif a7 == 1:
        return '310 - Scanning for Vulnerable Software'
    elif a8 == 1:
        return '153 - Input Data Manipulation'
    elif a9 == 1:
        return '248 - Command Injection'
    elif a10 == 1:
        return '274 - HTTP Verb Tampering'
    elif a11 == 1:
        return '194 - Fake the Source of Data'
    elif a12 == 1:
        return '34 - HTTP Response Splitting'
    elif a13 == 1:
        return '33 - HTTP Request Smuggling'
    return 'Error'

        
df_capec['label'] = df_capec.apply(lambda x: count_duplicate(x['000 - Normal'],x['272 - Protocol Manipulation'], x['242 - Code Injection'],x['88 - OS Command Injection'],x['126 - Path Traversal'],x['66 - SQL Injection'],x['16 - Dictionary-based Password Attack'],x['310 - Scanning for Vulnerable Software'],x['153 - Input Data Manipulation'],x['248 - Command Injection'],x['274 - HTTP Verb Tampering'],x['194 - Fake the Source of Data'],x['34 - HTTP Response Splitting'],x['33 - HTTP Request Smuggling']), axis = 1)
```

```python
df_capec.head()
```

```python
df_training = df_capec[(df_capec['label'] != '33 - HTTP Request Smuggling') & (df_capec['label'] != '248 - Command Injection') & (df_capec['label'] != '16 - Dictionary-based Password Attack') & (df_capec['label'] != '274 - HTTP Verb Tampering') & (df_capec['label'] != '88 - OS Command Injection')]
```
